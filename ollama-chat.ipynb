```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
```

```python
llm = ChatOllama(model="llama3.2:1b")
```

```python
prompt = PromptTemplate(
    input_variables=["country"],
    template="What is the capital of {country}? \
        Before extracting the answer, verify the information and make sure your answer is correct. \
        then just give the answer in a word., no explanation.",
)
```

```python
chain = (prompt | llm).invoke({"country": "Iran"})
print(chain.content)
```
